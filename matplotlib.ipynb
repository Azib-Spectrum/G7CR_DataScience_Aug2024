```python
# Categorical Data
# Measurable Data/Measures

# Guidelines:
# Bar Chart
    # Comparative Analysis
    # x-axis as categorical and y-axis as measurable
# Pie Chart
    # Composition Analysis
    # one categorical and one measurable value
    # if categories are more than 5-7, switch to using a bar chart
# Line Chart
    # Trend Analysis
    # x-axis is always date/time, y-axis is measurable
    # eg: stock market analysis, heart rate monitor,
# Scatter plot
    # Clustering Analysis
    # both x-axis and y-axis are measurable
# Histogram
    # Frequency Analysis
    # one axis, y-axis is always count/frequency/density
```

```python
time_of_day=['Dawn','Morning','Brunch','Noon','Afternoon','Dusk','Evening','Midnight']
sales_day1=[50,78,42,20,57,72,102,45]
sales_day2=[45,85,33,24,45,70,85,40]
sales_day3=[40,80,25,15,38,66,82,56]
```

```python
plt.plot(time_of_day,sales_day1,color="k",label="Day 1",linestyle="--",marker="*")
plt.bar(time_of_day,sales_day2)
```

```python
plt.figure(figsize=(12,4))
plt.plot(time_of_day,sales_day1,color="pink",label="Day 1",linestyle="--",marker="*")
plt.plot(time_of_day,sales_day2,color='b',label="Day 2",linestyle=":",marker="*")
plt.plot(time_of_day,sales_day3,color='#65c704',label="Day 3",linestyle="-.",marker="*")
plt.legend(loc="lower right")
plt.xticks(rotation=40)
plt.yticks(np.arange(10,111,10))
plt.xlabel("Time of the day",fontsize=11)
plt.ylabel("Number of cups of tea sold")
plt.title("Sales Trend for tea over the past 3 days")
plt.grid(linestyle='--',axis="y")
plt.show()
```
